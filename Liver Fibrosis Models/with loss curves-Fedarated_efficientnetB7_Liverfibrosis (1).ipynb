```python
# Colab-ready cell: FedAvg with EfficientNet-B7 + Central Test Evaluation + Best-model tracking + Per-client loss curves
# Paste into Colab. Edit BASE_DIR and TEST_DIR.

# -------------------------
# 0) Mount Drive (if not already)
# -------------------------
from google.colab import drive
drive.mount('/content/drive')

# -------------------------
# 1) USER SETTINGS (EDIT)
# -------------------------
BASE_DIR = '/content/drive/MyDrive/Colab Notebooks/liver_fibrosis/Dataset'   # <-- client folders (each immediate subfolder is a client)
TEST_DIR = '/content/drive/MyDrive/Colab Notebooks/liver_fibrosis/test'     # <-- central test dataset (ImageFolder-style)
SAVE_PATH = '/content/drive/MyDrive/Colab Notebooks/liver_fibrosis/global_efficientnetb7.pth'
ROUND_SAVE_PREFIX = '/content/drive/MyDrive/Colab Notebooks/liver_fibrosis/global_efficientnetb7'

ROUNDS = 7
LOCAL_EPOCHS = 2
BATCH_SIZE = 2
IMG_SIZE = 600
LR = 5e-5
WEIGHT_DECAY = 1e-4
NUM_WORKERS = 2

import torch
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Using device:", DEVICE)

# -------------------------
# 2) Libraries
# -------------------------
import os, gc, glob, warnings, copy
from collections import OrderedDict
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms, models
from PIL import Image, UnidentifiedImageError
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, classification_report, confusion_matrix

warnings.filterwarnings("ignore")
sns.set()

# -------------------------
# 3) Helper: allowed extensions and file finding
# -------------------------
ALLOWED_EXT = {'.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff', '.webp', '.ppm', '.pgm'}

def find_images_in_dir(root_dir):
    """Recursively find files with allowed extensions under root_dir."""
    files = []
    for ext in ALLOWED_EXT:
        files.extend(glob.glob(os.path.join(root_dir, '**', f'*{ext}'), recursive=True))
        files.extend(glob.glob(os.path.join(root_dir, '**', f'*{ext.upper()}'), recursive=True))
    files = sorted(set(files))
    return files

# -------------------------
# 4) Robust CustomImageFolder dataset (skips missing/corrupt files)
# -------------------------
class CustomImageFolder(Dataset):
    """
    Similar to torchvision.datasets.ImageFolder but:
    - Builds samples by scanning class subfolders recursively.
    - Filters out missing files at creation time.
    - If an image fails to open at __getitem__, it will attempt a few retries and remove the bad sample.
    """
    def __init__(self, root, transform=None, max_open_tries=3):
        self.root = root
        self.transform = transform
        self.max_open_tries = max_open_tries
        self.classes = []
        self.class_to_idx = {}
        self.samples = []  # list of (filepath, class_idx)
        self._make_dataset()

    def _make_dataset(self):
        entries = sorted(os.listdir(self.root)) if os.path.exists(self.root) else []
        for name in entries:
            p = os.path.join(self.root, name)
            if os.path.isdir(p):
                imgs = find_images_in_dir(p)
                # filter only existing files (defensive)
                valid_imgs = [fp for fp in imgs if os.path.exists(fp)]
                if len(valid_imgs) > 0:
                    idx = len(self.classes)
                    self.classes.append(name)
                    self.class_to_idx[name] = idx
                    for fp in valid_imgs:
                        self.samples.append((fp, idx))
                else:
                    print(f"Warning: class folder '{name}' under '{self.root}' has no valid image files. Skipping class.")
        if len(self.classes) == 0:
            imgs = find_images_in_dir(self.root)
            valid_imgs = [fp for fp in imgs if os.path.exists(fp)]
            if len(valid_imgs) > 0:
                self.classes = ['all']
                self.class_to_idx = {'all': 0}
                for fp in valid_imgs:
                    self.samples.append((fp, 0))
        # Final defensive filter (remove any non-existing paths)
        initial_len = len(self.samples)
        self.samples = [(fp, t) for fp, t in self.samples if os.path.exists(fp)]
        removed = initial_len - len(self.samples)
        if removed > 0:
            print(f"Warning: removed {removed} non-existing files from samples during dataset build.")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        # Defensive: allow retries and remove bad samples on-the-fly
        if len(self.samples) == 0:
            raise IndexError("Dataset contains no valid image samples.")

        # keep an index within bounds
        idx = idx % len(self.samples)
        tries = 0
        while tries < self.max_open_tries:
            path, target = self.samples[idx]
            try:
                img = Image.open(path).convert('RGB')
                if self.transform:
                    img = self.transform(img)
                return img, target
            except (FileNotFoundError, UnidentifiedImageError, OSError) as e:
                # Bad file — log and remove from samples, then try next
                print(f"Warning: failed to open image '{path}'. Removing from dataset. Error: {e}")
                try:
                    # remove the bad sample
                    self.samples.pop(idx)
                except Exception:
                    pass
                if len(self.samples) == 0:
                    raise IndexError("All samples failed to open; dataset is empty after filtering.")
                # do not increment idx — we want to try the next item that took this index
                tries += 1
        # If we exhausted retries, raise
        raise RuntimeError(f"Failed to load a valid image after {self.max_open_tries} attempts.")

# -------------------------
# 5) Transforms (your originals)
# -------------------------
def get_transforms(img_size, train_mode=True):
    if train_mode:
        return transforms.Compose([
            transforms.RandomResizedCrop(img_size, scale=(0.9, 1.0)),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomRotation(degrees=10),
            transforms.RandomAffine(0, translate=(0.05,0.05), scale=(0.95,1.05)),
            transforms.RandomAdjustSharpness(sharpness_factor=1.5, p=0.3),
            transforms.RandomAutocontrast(p=0.3),
            transforms.ToTensor(),
            transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
        ])
    else:
        return transforms.Compose([
            transforms.Resize((img_size, img_size)),
            transforms.ToTensor(),
            transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
        ])

# -------------------------
# 6) Model loader (EfficientNet-B7)
# -------------------------
def load_model_core(num_classes, pretrained=True):
    """
    Load EfficientNet-B7 and replace classifier. Returns model (cpu).
    Robust to different torchvision API versions.
    """
    try:
        # Preferred modern API
        weights = models.EfficientNet_B7_Weights.IMAGENET1K_V1 if pretrained else None
        model = models.efficientnet_b7(weights=weights)
    except Exception:
        try:
            # Older fallback (string-style)
            model = models.efficientnet_b7(weights="IMAGENET1K_V1" if pretrained else None)
        except Exception:
            # Ultimate fallback: older torchvision versions
            model = models.efficientnet_b7(pretrained=pretrained)

    # Replace classifier safely (different versions use model.classifier as Sequential)
    try:
        # Most torchvision implementations: classifier is nn.Sequential([Dropout, Linear])
        in_features = model.classifier[1].in_features
        model.classifier[1] = nn.Linear(in_features, num_classes)
    except Exception:
        # Find last Linear layer in classifier sequential
        try:
            if hasattr(model, 'classifier') and isinstance(model.classifier, nn.Sequential):
                last_linear_idx = None
                for i, layer in enumerate(model.classifier):
                    if isinstance(layer, nn.Linear):
                        last_linear_idx = i
                if last_linear_idx is not None:
                    in_f = list(model.classifier[last_linear_idx].parameters())[0].shape[1]
                    seq = list(model.classifier)
                    seq[last_linear_idx] = nn.Linear(in_f, num_classes)
                    model.classifier = nn.Sequential(*seq)
            elif hasattr(model, 'classifier') and isinstance(model.classifier, nn.Linear):
                in_f = list(model.classifier.parameters())[0].shape[1]
                model.classifier = nn.Linear(in_f, num_classes)
        except Exception:
            # if everything fails, create a simple final linear replacing attribute
            try:
                model.classifier = nn.Linear(2560, num_classes)  # 2560 is EfficientNet-B7 final feature size in many builds
            except Exception:
                pass

    return model

# -------------------------
# 7) Helper to evaluate an arbitrary state_dict on a given test dataset (used for central testing)
# -------------------------
def evaluate_state_on_test(state_dict, test_ds, img_size=600, batch_size=8, device='cuda'):
    """
    Build a model for test_ds.classes, merge state_dict where shapes match, and evaluate.
    Returns metrics dict similar to evaluate_global_model but without plotting confusion matrix.
    """
    if len(test_ds) == 0:
        return None
    test_loader = DataLoader(test_ds, batch_size=batch_size, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
    num_classes = len(test_ds.classes)
    model = load_model_core(num_classes, pretrained=False).to(device)

    # merge states where shapes match
    model_state = model.state_dict()
    for k, v in state_dict.items():
        if k in model_state and model_state[k].shape == v.shape:
            model_state[k] = v.clone()
    model.load_state_dict(model_state)

    model.eval()
    y_true = []
    y_pred = []
    total_loss = 0.0
    total_samples = 0
    criterion = nn.CrossEntropyLoss()
    with torch.no_grad():
        if device and isinstance(device, torch.device) and device.type == 'cuda':
            autocast = torch.cuda.amp.autocast
        else:
            from contextlib import nullcontext
            autocast = nullcontext
        with autocast():
            for imgs, labels in test_loader:
                imgs = imgs.to(device)
                labels = labels.to(device)
                outputs = model(imgs)
                loss = criterion(outputs, labels)
                preds = outputs.argmax(1)
                y_true.extend(labels.cpu().numpy())
                y_pred.extend(preds.cpu().numpy())
                total_loss += loss.item() * imgs.size(0)
                total_samples += imgs.size(0)

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='macro', zero_division=0)
    rec = recall_score(y_true, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)
    avg_loss = total_loss / max(1, total_samples)

    metrics = {
        'test_samples': len(test_ds),
        'accuracy': acc,
        'precision_macro': prec,
        'recall_macro': rec,
        'f1_macro': f1,
        'loss': avg_loss,
        'classes': test_ds.classes,
        'y_true': y_true,
        'y_pred': y_pred
    }
    return metrics

# -------------------------
# 8) ClientTrainer using CustomImageFolder (returns per-epoch train & test losses)
# -------------------------
class ClientTrainer:
    def __init__(self, client_path, img_size=600, batch_size=2, device='cpu', num_workers=2):
        self.client_path = client_path
        self.img_size = img_size
        self.batch_size = batch_size
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')
        self.num_workers = num_workers
        self._prepare()

    def _prepare(self):
        self.train_transform = get_transforms(self.img_size, train_mode=True)
        self.eval_transform = get_transforms(self.img_size, train_mode=False)
        self.train_dataset = CustomImageFolder(self.client_path, transform=self.train_transform)
        self.num_samples = len(self.train_dataset)
        self.classes = self.train_dataset.classes
        self.num_classes = len(self.classes)
        if self.num_samples == 0:
            print(f"Warning: client folder '{self.client_path}' has no valid images. This client will be skipped.")
            self.sampler = None
            self.train_loader = None
            return

        labels = [s[1] for s in self.train_dataset.samples]
        class_counts = np.bincount(labels, minlength=max(1, self.num_classes))
        class_counts = np.where(class_counts == 0, 1, class_counts)
        class_weights = 1.0 / class_counts
        samples_weight = [class_weights[l] for l in labels]
        self.sampler = WeightedRandomSampler(samples_weight, num_samples=self.num_samples, replacement=True)
        self.train_loader = DataLoader(self.train_dataset,
                                       batch_size=self.batch_size,
                                       sampler=self.sampler,
                                       num_workers=self.num_workers,
                                       pin_memory=True)

    def local_train(self, base_state_dict, epochs=1, lr=5e-5, weight_decay=1e-4, central_test_ds=None, test_batch=8):
        """
        Performs local training for `epochs`. After each epoch, evaluates on central_test_ds (if provided)
        Returns: final_state_dict, n_samples, avg_loss_over_all_epochs, train_losses_list, test_losses_list, test_acc_list
        """
        if self.num_samples == 0 or self.train_loader is None:
            # return empty lists so federated loop can handle it
            return base_state_dict, 0, 0.0, [], [], []

        model = load_model_core(max(self.num_classes, 2), pretrained=False).to(self.device)

        # load compatible weights from base_state_dict
        model_state = model.state_dict()
        for k, v in base_state_dict.items():
            if k in model_state and model_state[k].shape == v.shape:
                model_state[k] = v.clone()
        model.load_state_dict(model_state)

        criterion = nn.CrossEntropyLoss(label_smoothing=0.05)
        optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
        scaler = torch.cuda.amp.GradScaler()

        model.train()
        total_instances = 0
        running_loss_total = 0.0

        train_losses = []
        test_losses = []
        test_accs = []

        for ep in range(epochs):
            gc.collect(); torch.cuda.empty_cache()
            epoch_loss_numer = 0.0
            epoch_seen = 0
            pbar = tqdm(self.train_loader, desc=f"[{os.path.basename(self.client_path)}] epoch {ep+1}/{epochs}")
            for imgs, labels in pbar:
                imgs = imgs.to(self.device)
                labels = labels.to(self.device)
                optimizer.zero_grad()
                with torch.cuda.amp.autocast():
                    outputs = model(imgs)
                    loss = criterion(outputs, labels)
                preds = outputs.argmax(1)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
                batch_sz = labels.size(0)
                epoch_loss_numer += loss.item() * batch_sz
                epoch_seen += batch_sz
                running_loss_total += loss.item() * batch_sz
                total_instances += batch_sz
                pbar.set_postfix({'epoch_loss': epoch_loss_numer / max(1, epoch_seen)})

            # epoch finished
            epoch_avg_loss = epoch_loss_numer / max(1, epoch_seen)
            train_losses.append(epoch_avg_loss)

            # evaluate on central test set (if provided)
            if central_test_ds is not None and len(central_test_ds) > 0:
                metrics = evaluate_state_on_test(model.cpu().state_dict(), central_test_ds, img_size=self.img_size, batch_size=test_batch, device=self.device)
                model.to(self.device)  # move back to device for further training
                if metrics is not None:
                    test_losses.append(metrics['loss'])
                    test_accs.append(metrics['accuracy'])
                    print(f" -> Client {os.path.basename(self.client_path)} epoch {ep+1} TEST acc: {metrics['accuracy']*100:.4f}% loss: {metrics['loss']:.6f}")
                else:
                    test_losses.append(None)
                    test_accs.append(None)
            else:
                test_losses.append(None)
                test_accs.append(None)

        avg_loss_all = running_loss_total / max(1, total_instances)
        # return CPU state dict for lightweight aggregation
        return model.cpu().state_dict(), self.num_samples, avg_loss_all, train_losses, test_losses, test_accs

# -------------------------
# 9) Fixed FedAvg aggregation (handles dtype mismatch)
# -------------------------
def fedavg_aggregate(global_state, client_states, client_samples):
    total_samples = sum(client_samples)
    if total_samples == 0:
        return global_state

    new_state = OrderedDict()
    float_keys = []
    for k, v in global_state.items():
        if torch.is_floating_point(v):
            new_state[k] = torch.zeros_like(v, dtype=torch.float32)
            float_keys.append(k)
        else:
            new_state[k] = v.clone()

    for state, n in zip(client_states, client_samples):
        if n == 0: continue
        w = float(n) / float(total_samples)
        for k in float_keys:
            if k in state:
                new_state[k] += state[k].to(dtype=torch.float32) * w

    for k in float_keys:
        orig = global_state[k]
        dtype = orig.dtype
        new_state[k] = new_state[k].to(dtype)

    return new_state

# -------------------------
# 10) Central test evaluator (keeps plotting capability)
# -------------------------
def evaluate_global_model(global_state, test_dir, img_size=600, batch_size=8, device='cuda'):
    test_ds = CustomImageFolder(test_dir, transform=get_transforms(img_size, train_mode=False))
    if len(test_ds) == 0:
        print("No images found in TEST_DIR:", test_dir)
        return None
    metrics = evaluate_state_on_test(global_state, test_ds, img_size=img_size, batch_size=batch_size, device=device)

    if metrics is None:
        return None

    print("\n=== TEST SET RESULTS ===")
    print(f"Test samples      : {len(test_ds)}")
    print(f"Accuracy          : {metrics['accuracy']*100:.4f}%")
    print(f"Macro Precision   : {metrics['precision_macro']:.4f}")
    print(f"Macro Recall      : {metrics['recall_macro']:.4f}")
    print(f"Macro F1-score    : {metrics['f1_macro']:.4f}\n")

    print("Classification Report:")
    try:
        print(classification_report(metrics['y_true'], metrics['y_pred'], target_names=test_ds.classes, zero_division=0))
    except Exception as e:
        print("Could not print classification report:", e)

    # Confusion matrix
    cm = confusion_matrix(metrics['y_true'], metrics['y_pred'])
    plt.figure(figsize=(6,6))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=test_ds.classes, yticklabels=test_ds.classes, cmap='Blues')
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix (Test Set)")
    plt.show()

    return metrics

# -------------------------
# 11) Federated loop (with robust client filtering & per-client curves)
# -------------------------
def run_federated(base_dir, test_dir=None, rounds=5, local_epochs=1, batch_size=2, img_size=600, lr=5e-5, weight_decay=1e-4, device='cuda', save_path=SAVE_PATH, round_save_prefix=ROUND_SAVE_PREFIX):
    device = torch.device(device if torch.cuda.is_available() else 'cpu')
    print("Federated device:", device)

    if not os.path.exists(base_dir):
        raise FileNotFoundError(f"BASE_DIR does not exist: {base_dir}\nDid you mount Drive and edit BASE_DIR correctly?")
    clients = []
    for name in sorted(os.listdir(base_dir)):
        p = os.path.join(base_dir, name)
        if os.path.isdir(p):
            clients.append((name, p))
    if len(clients) == 0:
        raise FileNotFoundError(f"No client subfolders found under BASE_DIR={base_dir}. Directory listing:\n{os.listdir(base_dir)}")

    trainers = []
    for name, path in clients:
        print(f"Scanning client '{name}' at: {path}")
        t = ClientTrainer(path, img_size=img_size, batch_size=batch_size, device=device, num_workers=NUM_WORKERS)
        if t.num_samples == 0:
            print(f" -> Skipping client '{name}' (no valid images found).")
        else:
            print(f" -> Found {t.num_samples} images across {t.num_classes} classes: {t.classes}")
            trainers.append((name, t))

    if len(trainers) == 0:
        sample_listing = []
        for root, dirs, files in os.walk(base_dir):
            sample_listing.extend(files[:10])
            if len(sample_listing) >= 10: break
        raise FileNotFoundError(f"No valid clients with images were found under {base_dir}.\nSample listing (first files found): {sample_listing}\nCheck extensions: {sorted(list(ALLOWED_EXT))}")

    # central test dataset (shared)
    test_ds = None
    if test_dir is not None and os.path.exists(test_dir):
        test_ds = CustomImageFolder(test_dir, transform=get_transforms(img_size, train_mode=False))
        print("Central TEST_DIR classes:", test_ds.classes if test_ds is not None else None)
    else:
        print("No TEST_DIR provided or does not exist. Per-epoch central test evaluations will be skipped.")

    # infer number of classes from first non-empty client (fallback if test_ds is empty)
    first_client_path = trainers[0][1].client_path
    sample_ds = CustomImageFolder(first_client_path, transform=get_transforms(img_size, train_mode=False))
    num_classes = len(sample_ds.classes) if len(sample_ds.classes)>0 else max(2, trainers[0][1].num_classes)
    print("Global class mapping inferred from first non-empty client:", sample_ds.classes)

    global_model = load_model_core(num_classes, pretrained=True)
    global_state = global_model.state_dict()

    # tracking
    train_loss_per_round = []
    test_loss_per_round = []
    test_f1_per_round = []
    best_f1 = -1.0
    best_state = None
    best_round = -1

    # per-client curve storage (dictionary: client_name -> lists)
    client_curves = {name: {'train_losses': [], 'test_losses': [], 'test_accs': []} for name, _ in trainers}

    for r in range(1, rounds+1):
        print(f"\n=== Federated Round {r}/{rounds} ===")
        client_states = []
        client_samples = []
        round_train_loss_numer = 0.0
        round_train_samples = 0
        for name, trainer in trainers:
            print(f"\n-> Training client: {name} | samples: {trainer.num_samples}")
            (state_dict, n, avg_loss,
             train_losses_list, test_losses_list, test_accs_list) = trainer.local_train(global_state,
                                                                                         epochs=local_epochs,
                                                                                         lr=lr,
                                                                                         weight_decay=weight_decay,
                                                                                         central_test_ds=test_ds,
                                                                                         test_batch=max(1, BATCH_SIZE*4))
            client_states.append(state_dict)
            client_samples.append(n)
            print(f"Client {name} done. samples={n}, avg_loss_over_epochs={avg_loss:.4f}")
            round_train_loss_numer += avg_loss * n
            round_train_samples += n

            # append per-client epoch lists to client_curves
            if len(train_losses_list) > 0:
                client_curves[name]['train_losses'].append(train_losses_list)
            else:
                client_curves[name]['train_losses'].append([])

            if len(test_losses_list) > 0:
                client_curves[name]['test_losses'].append(test_losses_list)
            else:
                client_curves[name]['test_losses'].append([])

            if len(test_accs_list) > 0:
                client_curves[name]['test_accs'].append(test_accs_list)
            else:
                client_curves[name]['test_accs'].append([])

        # weighted average train loss for this round
        round_train_loss = round_train_loss_numer / max(1, round_train_samples)
        train_loss_per_round.append(round_train_loss)

        new_global_state = fedavg_aggregate(global_state, client_states, client_samples)
        global_state = new_global_state
        print(f"Round {r} aggregated.")

        round_save = f"{round_save_prefix}_round{r}.pth"
        torch.save({'model_state': global_state, 'num_classes': num_classes}, round_save)
        print("Saved intermediate model:", round_save)

        # Evaluate immediately on test set and keep best model by macro F1
        if test_ds is not None:
            print("\nEvaluating aggregated global model on TEST_DIR...")
            metrics = evaluate_state_on_test(global_state, test_ds, img_size=img_size, batch_size=max(1, BATCH_SIZE*4), device=device)
            if metrics is not None:
                test_loss_per_round.append(metrics['loss'])
                test_f1_per_round.append(metrics['f1_macro'])
                current_f1 = metrics['f1_macro']
                print(f"Aggregated model TEST F1: {current_f1:.4f} | Accuracy: {metrics['accuracy']*100:.4f}%")
                if current_f1 > best_f1:
                    best_f1 = current_f1
                    best_state = copy.deepcopy(global_state)
                    best_round = r
                    # save best
                    torch.save({'model_state': best_state, 'num_classes': num_classes}, f"{round_save_prefix}_best.pth")
                    print(f"New best model saved (round {r}) with F1={best_f1:.4f}")
        else:
            print("No TEST_DIR provided or it does not exist. Skipping per-round evaluation.")

    # final save
    torch.save({'model_state': global_state, 'num_classes': num_classes}, save_path)
    print("Saved final global model:", save_path)

    # Load best model (if any) and run final evaluation
    if best_state is not None and test_ds is not None:
        print(f"\nLoading best model from round {best_round} (F1={best_f1:.4f}) for final evaluation...")
        final_metrics = evaluate_state_on_test(best_state, test_ds, img_size=img_size, batch_size=max(1, BATCH_SIZE*4), device=device)
        print("Final evaluation completed on best model.")
    else:
        print("\nNo best model found (no TEST_DIR or no evaluations ran). Running evaluation on final global state if test_dir exists.")
        if test_ds is not None:
            final_metrics = evaluate_state_on_test(global_state, test_ds, img_size=img_size, batch_size=max(1, BATCH_SIZE*4), device=device)
        else:
            final_metrics = None

    # Plot aggregated loss and F1 curves
    rounds_x = list(range(1, len(train_loss_per_round)+1))
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(rounds_x, train_loss_per_round, marker='o')
    plt.title('Weighted avg client train loss per round')
    plt.xlabel('Round')
    plt.ylabel('Train loss')

    plt.subplot(1,2,2)
    if len(test_loss_per_round) > 0:
        plt.plot(list(range(1, len(test_loss_per_round)+1)), test_loss_per_round, marker='o', label='test loss')
    if len(test_f1_per_round) > 0:
        plt.plot(list(range(1, len(test_f1_per_round)+1)), test_f1_per_round, marker='x', label='test F1')
    plt.title('Test loss and F1 per round')
    plt.xlabel('Round')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Plot per-client epoch curves (for all rounds -> flatten epoch lists)
    # client_curves[name]['train_losses'] is list per round, each entry is list of local-epoch losses
    n_clients = len(client_curves)
    cols = min(3, n_clients)
    rows = (n_clients + cols - 1) // cols
    plt.figure(figsize=(cols*5, rows*4))
    idx = 1
    for cname, data in client_curves.items():
        # flatten epochs across rounds
        flat_train = [x for round_list in data['train_losses'] for x in round_list]
        flat_test = [x for round_list in data['test_losses'] for x in round_list]
        plt.subplot(rows, cols, idx)
        if len(flat_train) > 0:
            plt.plot(list(range(1, len(flat_train)+1)), flat_train, marker='o', label='train loss')
        if len(flat_test) > 0 and any([t is not None for t in flat_test]):
            # convert None to NaN so plotting works
            flat_test_plot = [t if t is not None else np.nan for t in flat_test]
            plt.plot(list(range(1, len(flat_test_plot)+1)), flat_test_plot, marker='x', label='test loss')
        plt.title(f"Client: {cname}")
        plt.xlabel("Local epoch (across rounds)")
        plt.ylabel("Loss")
        plt.legend()
        idx += 1
    plt.tight_layout()
    plt.show()

    return {
        'train_loss_per_round': train_loss_per_round,
        'test_loss_per_round': test_loss_per_round,
        'test_f1_per_round': test_f1_per_round,
        'best_round': best_round,
        'best_f1': best_f1,
        'final_metrics': final_metrics,
        'client_curves': client_curves
    }

# -------------------------
# 12) Run
# -------------------------
if __name__ == "__main__":
    results = run_federated(base_dir=BASE_DIR,
                            test_dir=TEST_DIR,
                            rounds=ROUNDS,
                            local_epochs=LOCAL_EPOCHS,
                            batch_size=BATCH_SIZE,
                            img_size=IMG_SIZE,
                            lr=LR,
                            weight_decay=WEIGHT_DECAY,
                            device='cuda',
                            save_path=SAVE_PATH,
                            round_save_prefix=ROUND_SAVE_PREFIX)

    print("\nRun finished. Summary:")
    print("Best round:", results['best_round'], " Best F1:", f"{results['best_f1']:.4f}")
    if results['final_metrics'] is not None:
        print("Final test accuracy: ", f"{results['final_metrics']['accuracy']*100:.4f}%")
```

```python
# Confusion matrix-only evaluator (paste into Colab cell)
import os, glob
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from torchvision import transforms
from torch.utils.data import DataLoader

# Edit if needed (will try to use TEST_DIR, SAVE_PATH, ROUND_SAVE_PREFIX from your notebook if present)
try:
    TEST_DIR  # if variable exists, use it
except NameError:
    TEST_DIR = '/content/drive/MyDrive/Colab Notebooks/liver_fibrosis/test'

try:
    SAVE_PATH
except NameError:
    SAVE_PATH = None

try:
    ROUND_SAVE_PREFIX
except NameError:
    ROUND_SAVE_PREFIX = None

NUM_WORKERS = globals().get('NUM_WORKERS', 2)
IMG_SIZE = globals().get('IMG_SIZE', 600)
BATCH_SIZE = max(1, globals().get('BATCH_SIZE', 8))
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def find_checkpoint():
    candidates = []
    if ROUND_SAVE_PREFIX:
        pbest = f"{ROUND_SAVE_PREFIX}_best.pth"
        if os.path.exists(pbest):
            return pbest
        # fallback to any round files
        candidates.extend(sorted(glob.glob(f"{ROUND_SAVE_PREFIX}_round*.pth")))
    if SAVE_PATH and os.path.exists(SAVE_PATH):
        candidates.append(SAVE_PATH)
    # fallback: any pth in current working dir (prefer newest)
    candidates.extend(sorted(glob.glob("*.pth"), key=os.path.getmtime, reverse=True))
    return candidates[0] if len(candidates) > 0 else None

ckpt_path = find_checkpoint()
if ckpt_path is None:
    raise FileNotFoundError("No checkpoint found. Please set ROUND_SAVE_PREFIX or SAVE_PATH, or place a .pth in the working dir.")

print("Using checkpoint:", ckpt_path)

ckpt = torch.load(ckpt_path, map_location='cpu')

# Determine model state and num_classes in checkpoint
if isinstance(ckpt, dict) and ('model_state' in ckpt or 'state_dict' in ckpt):
    state = ckpt.get('model_state', ckpt.get('state_dict'))
    num_classes_ckpt = ckpt.get('num_classes', None)
else:
    # assume ckpt is a raw state_dict
    state = ckpt
    num_classes_ckpt = None

# Build test dataset
test_ds = CustomImageFolder(TEST_DIR, transform=get_transforms(IMG_SIZE, train_mode=False))
if len(test_ds) == 0:
    raise FileNotFoundError(f"No test images found under TEST_DIR: {TEST_DIR}")

# If checkpoint didn't store num_classes, infer from test_ds
num_classes = num_classes_ckpt if (num_classes_ckpt is not None) else len(test_ds.classes)
print("Num classes (used):", num_classes, "Test classes:", test_ds.classes)

# Load model and merge compatible keys (robust)
model = load_model_core(num_classes, pretrained=False)
model_state = model.state_dict()
for k, v in state.items():
    if k in model_state and model_state[k].shape == v.shape:
        model_state[k] = v.clone()
model.load_state_dict(model_state)
model.to(DEVICE)
model.eval()

# DataLoader
test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

# Run inference
y_true = []
y_pred = []
with torch.no_grad():
    for imgs, labels in test_loader:
        imgs = imgs.to(DEVICE)
        labels = labels.to(DEVICE)
        outputs = model(imgs)
        preds = outputs.argmax(1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())

# Metrics & confusion matrix
acc = accuracy_score(y_true, y_pred)
print(f"\nTest samples: {len(test_ds)}")
print(f"Accuracy: {acc*100:.4f}%\n")
print("Classification report:")
try:
    print(classification_report(y_true, y_pred, target_names=test_ds.classes, zero_division=0))
except Exception as e:
    print("Could not produce classification report:", e)

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=test_ds.classes, yticklabels=test_ds.classes, cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"Confusion Matrix ({os.path.basename(ckpt_path)})")
plt.show()
```
